```python
# 필요한 라이브러리 임포트
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 한글 폰트를 설정하기 위한 라이브러리 임포트
from matplotlib import rc

# 한글 폰트 설정 (예: 맑은 고딕 사용)
rc('font', family='Malgun Gothic')

# txt 파일에서 텍스트 데이터 읽어오기
file_path = './ingredients.txt'  # .txt 파일의 실제 경로를 입력하세요
with open(file_path, 'r', encoding='utf-8') as file:
    text1 = file.read()
```

```python
# 필요한 라이브러리 임포트
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 한글 폰트를 설정하기 위한 라이브러리 임포트
from matplotlib import rc

# 한글 폰트 설정 (예: 맑은 고딕 사용)
rc('font', family='Malgun Gothic')

# txt 파일에서 텍스트 데이터 읽어오기
file_path = './health.txt'  # .txt 파일의 실제 경로를 입력하세요
with open(file_path, 'r', encoding='utf-8') as file:
    text2 = file.read()
```

```python
from collections import Counter

# 텍스트를 단어별로 분리
words = text1.split()
print(f"총 단어 수: {len(words)}")

# 단어 빈도 계산
word_counts1 = Counter(words)
print(word_counts1.most_common(10))  # 가장 많이 등장하는 단어 10개
```

```python
from konlpy.tag import Okt
from collections import Counter

# 불용어 리스트 정의
stopwords = ['(이)가','들어있습니다.', '이']

# KoNLPy의 Okt 객체를 사용하여 형태소 분석
okt = Okt()

# 텍스트를 토큰화 및 명사 추출
tokens = okt.nouns(text1) #ingredients

# 불용어 제거
filtered_tokens = [word for word in tokens if word not in stopwords]

# 단어 빈도수 계산
word_counts1 = Counter(filtered_tokens)

# 상위 10개의 단어 출력
top_10 = word_counts1.most_common(10)
print("단어 빈도수 상위 10개:")
for word, count in top_10:
    print(f"{word}: {count}")
```

```python
# 워드 클라우드 생성
wordcloud = WordCloud(font_path='C:/Windows/Fonts/malgun.ttf', 
                       background_color="white", 
                       width=800, 
                       height=400).generate_from_frequencies(word_counts1)

# 워드 클라우드 시각화
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("<워드 클라우드>", pad=20)
plt.show()
```

```python
from konlpy.tag import Okt
from collections import Counter

# 불용어 리스트 정의
stopwords = ['수', '것', '것이', '합니다.', '있습니다.', '중요합니다.', '위해', '통해', '또한', '도움이','적절한','유지하는','증상을','위해서는','있는', '데', '것도','좋습니다.','것을','다양한']

# KoNLPy의 Okt 객체를 사용하여 형태소 분석
okt = Okt()

# 텍스트를 토큰화 및 명사 추출
tokens = okt.nouns(text2) #health care

# 불용어 제거
filtered_tokens = [word for word in tokens if word not in stopwords]

# 단어 빈도수 계산
word_counts2 = Counter(filtered_tokens)

# 상위 10개의 단어 출력
top_10 = word_counts2.most_common(10)
print("단어 빈도수 상위 10개:")
for word, count in top_10:
    print(f"{word}: {count}")
```

```python
# 워드 클라우드 생성
wordcloud = WordCloud(font_path='C:/Windows/Fonts/malgun.ttf', 
                       background_color="white", 
                       width=800, 
                       height=400).generate_from_frequencies(word_counts2)

# 워드 클라우드 시각화
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("<워드 클라우드>", pad=20)
plt.show()
```

```python
# 상위 10개의 단어를 추출하여 그래프에 표시
top_10 = word_counts1.most_common(10)
words, counts = zip(*top_10)  # 단어와 빈도를 분리

# 그래프 생성
plt.figure(figsize=(10, 5))
plt.bar(words, counts, color='skyblue')
plt.xlabel("단어", fontsize=12)
plt.ylabel("빈도수", fontsize=12)
plt.title("어휘 사용 빈도 그래프", pad=20, fontsize=15)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
